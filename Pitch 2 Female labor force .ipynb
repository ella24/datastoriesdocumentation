# Calculate the GDP pc of 1960 inflated to 2019.


What is mean GDP pc of UMICs and LMICs and LICs today (or 2019)? 
The GDP PC series for OCED/HICs (those rich today) in the past, inflated to 2019 price levels (based on CPI). So the following data:
GDP pc in nominal terms for OECD/HICs for years X-Y (maybe 1950-1970?) in local current units (LCU)
Inflators for each year to 2019 for each of these countries.  So nominal value X inflator = 2019 value in LCUs
Exchange rate (one per country) to convert LCU into USD in 2019.
Then we can look at which developing countries today match to what (Sweden 1950?).

```python
#1970 Nominal GDP for HIC

df_GDPpc= wb.data.DataFrame(['NY.GDP.PCAP.CN'],["ABW","AND","ARE","ATG","AUS","AUT","BEL","BHR","BHS",
"BMU","BRB","BRN","CAN","CHE","CHI","CHL","CUW","CYM", 
"CYP","CZE","DEU","DNK","ESP","EST","FIN","FRA","FRO",
"GBR","GIB","GRC","GRL","GUM","HKG","HRV","HUN","IMN",
"IRL","ISL","ISR","ITA","JPN","KNA","KOR","KWT","LIE",
"LTU","LUX","LVA","MAC","MAF",
"MCO","MLT","MNP","MUS","NCL",
"NLD","NOR","NRU","NZL","OMN",
"PAN","PLW","POL","PRI","PRT",
"PYF","QAT","ROU","SAU","SGP",
"SMR","SVK","SVN","SWE","SXM","SYC",
"TCA",
"TTO",
"URY","USA","VGB","VIR"],time=(1960),labels=True)

df_GDPpc.head()
```

#Calculatin ratio of women vs men in vulnerable employment. 
 

(%w-%m)/%m = X

Where %w is percent for women.  m for men.

 

X will be negative to positive.

Example:

Women 20%, Men 40%, X=-1 (women much less likely than men)

Women 20%, Men 20%, X=0 (parity)

Women 40%, Men 20%, X=1 (women much more likely than men)


This could be plotted against GDP per cap (are working women more likely to be vulnerable workers than working men in low/high income countries). Or against average vulnerable employment % (are working women in countries with more vulnerable employment more likely to be vulnerable workers than working men). The two likely show similar things (since GDP pc and vulnerabile work are correlated)

```python
#Vulnerable employment, female (% of female employment) (modeled ILO estimate), male
vulnerable_df['ratio_women_men_vulnerable_employment']= (vulnerable_df['SL.EMP.VULN.FE.ZS']-vulnerable_df['SL.EMP.VULN.MA.ZS'])/vulnerable_df['SL.EMP.VULN.MA.ZS']
vulnerable_df.head()
```

```python
vulnerable_df.to_csv('vulnerableemploymentdata.csv', index=True)
```
